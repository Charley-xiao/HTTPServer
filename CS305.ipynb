```python
from Crypto.PublicKey import RSA
from Crypto.Random import get_random_bytes
from Crypto.Cipher import PKCS1_OAEP
from Crypto.Cipher import Salsa20
import base64
import requests

def generate_key():
    # Generate a random 256-bit (32-byte) key
    return get_random_bytes(32)

def save_key_to_file(key, filename):
    with open(filename, 'wb') as file:
        file.write(key)

def load_key_from_file(filename):
    with open(filename, 'rb') as file:
        return file.read()

def encrypt(message, key):
    cipher = Salsa20.new(key)
    ciphertext = cipher.encrypt(message.encode())
    return ciphertext

def decrypt(ciphertext, key):
    cipher = Salsa20.new(key)
    decrypted_message = cipher.decrypt(ciphertext).decode()
    return decrypted_message

# request public key from server
url = 'http://127.0.0.1:8080/encrypt?request=public_key'
headers={}
r=requests.get(url=url, headers=headers)
print(r)
public_key = r.content
save_key_to_file(public_key, './data/encryption/client/public_key.pem')

# generate Salsa20 key and encrypt it with RSA public key
Salsa20_key = generate_key()
save_key_to_file(Salsa20_key, './data/encryption/client/Salsa20_key.pem')
cipher_RSA = PKCS1_OAEP.new(RSA.importKey(public_key))
encrypted_Salsa20_key = cipher_RSA.encrypt(Salsa20_key)
encoded_Salsa20_key = base64.b64encode(encrypted_Salsa20_key).decode('utf-8')
print(encrypted_Salsa20_key)

# send encrypted Salsa20 key to server
url = 'http://127.0.0.1:8080/encrypt?response=encrypted'
headers={}
data={f'Encrypted key: {encoded_Salsa20_key}'}

r=requests.post(url=url, headers=headers, data=data)
print(r)

# send encrypted file to server
url = 'http://127.0.0.1:8080/encrypt?file=a.txt'
headers={}
r = requests.get(url=url, headers=headers)
print(r)
print(r.content.decode())
```
